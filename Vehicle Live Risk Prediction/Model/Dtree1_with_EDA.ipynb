```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Downloads/Vehicle Risk Prediction Dataset.csv")
```

```python
inputs = df.drop('Risk_Score',axis='columns')
target=df['Risk_Score']
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'inputs' is your DataFrame

le_visibility = LabelEncoder()
le_road_surface_conditions = LabelEncoder()
le_weather = LabelEncoder()
le_traffic_density = LabelEncoder()
le_road_hazards = LabelEncoder()
le_time_of_day = LabelEncoder()
le_fatigue_level = LabelEncoder()
le_medical_condition = LabelEncoder()
le_speeding = LabelEncoder()

inputs['visibility_n'] = le_visibility.fit_transform(inputs['Visibility'])
inputs['road_surface_conditions_n'] = le_road_surface_conditions.fit_transform(inputs['Road_Surface_Conditions'])
inputs['weather_n'] = le_weather.fit_transform(inputs['Weather'])
inputs['traffic_density_n'] = le_traffic_density.fit_transform(inputs['Traffic_Density'])
inputs['road_hazards_n'] = le_road_hazards.fit_transform(inputs['Road_Hazards'])
inputs['time_of_day_n'] = le_time_of_day.fit_transform(inputs['Time_of_Day'])
inputs['fatigue_level_n'] = le_fatigue_level.fit_transform(inputs['Fatigue_Level'])
inputs['medical_condition_n'] = le_medical_condition.fit_transform(inputs['Medical_Condition'])
inputs['speeding_n'] = le_speeding.fit_transform(inputs['Speeding'])
inputsinputs = df.drop('salary_more_then_100k',axis='columns')
```

```python

inputs_n= inputs.drop(['Light_Conditions' ,	'Road_Type' 	,'Landscape' ,'Visibility', 'Road_Surface_Conditions', 'Weather', 'Traffic_Density', 'Road_Hazards', 'Time_of_Day', 'Fatigue_Level', 'Medical_Condition', 'Speeding'],axis='columns')
inputs_n
```

```python

# Assuming 'target' is your original target column
target_n = target.apply(lambda x: 0 if x < 50 else 1)

# 1 means less than 50 low risk
# 2 means very high
# 0 means medium risk
```

```python


from sklearn import tree

model = tree.DecisionTreeClassifier(criterion='entropy', min_samples_split=10,min_samples_leaf=5, max_features='sqrt', random_state=42)

model.fit(inputs_n, target_n)
```

```python
model.score(inputs_n, target_n)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Pairplot numerical columns ke liye
sns.pairplot(df, hue='Risk_Score', palette='viridis')
plt.suptitle("Pairplot of Numerical Columns", y=1.02)
plt.show()

# Correlation heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Boxplots for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=col, y='Risk_Score', data=df)
    plt.title(f"Boxplot of {col} vs Risk_Score")
    plt.show()

# Violin plots for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=col, y='Risk_Score', data=df)
    plt.title(f"Violin Plot of {col} vs Risk_Score")
    plt.show()
```
